```python

# K-means 클러스터링
def get_kmeans_metric(k):
    kmeans = KMeans(n_clusters=k, n_init=10)
    kmeans.fit(tfidf_matrix)

    # 클러스터링 결과 확인
    labels = kmeans.labels_
    centroids = kmeans.cluster_centers_
    print("클러스터 레이블:", labels)
    print("중심점 위치:", centroids)

    # 응집도 계산
    inertia = kmeans.inertia_
    print("응집도:", inertia)

    # 분산 계산
    distances = kmeans.transform(tfidf_matrix)
    variance = np.mean(np.min(distances, axis=1))
    print("분산:", variance)

    # 실루엣 계수 계산
    silhouette_avg = silhouette_score(tfidf_matrix, labels)
    print("실루엣 계수:", silhouette_avg)
```

```python
# K 값의 범위 설정
k_values = range(1, 40)

# 응집도와 기울기를 저장할 리스트
inertias = []
gradients = []

# 각 K 값에 대해 K-Means 클러스터링 수행 및 응집도 계산
for k in k_values:
    kmeans = KMeans(n_clusters=k, n_init='auto')
    kmeans.fit(tfidf_matrix)
    inertia = kmeans.inertia_
    inertias.append(inertia)
    
# 기울기 계산
for i in range(1, len(inertias)):
    gradient = inertias[i-1] - inertias[i]
    gradients.append(gradient)

# 엘보우 그래프 그리기
plt.plot(k_values, inertias, 'bo-')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Inertia')
plt.title('Elbow Method')
plt.show()

# 기울기 값 출력
print("Gradients:")
for k, gradient in zip(k_values[1:], gradients):
    print(f"K={k}: {gradient}")
```

```python
# K 값의 범위 설정
k_values = range(2, 40)

# 실루엣 점수를 저장할 리스트
silhouette_scores = []

# 각 K 값에 대해 K-Means 클러스터링 수행 및 실루엣 점수 계산
for k in k_values:
    kmeans = KMeans(n_clusters=k, n_init='auto')
    kmeans.fit(tfidf_matrix)
    labels = kmeans.labels_
    score = silhouette_score(tfidf_matrix, labels)
    silhouette_scores.append(score)

# 실루엣 점수가 가장 높은 K 값을 선택
best_k = np.argmax(silhouette_scores) + 2

print("Best K value:", best_k)


# 실루엣 그래프 그리기
plt.plot(k_values, silhouette_scores, 'bo-')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('silhouette_scores')
plt.title('silhouette_scores Graph')
plt.show()
```

```python
import numpy as np
from sklearn.cluster import KMeans
from sklearn.preprocessing import normalize
from sklearn.metrics.pairwise import cosine_distances

class CosineKMeans(KMeans):
    def __init__(self, n_clusters=8, *, init='k-means++', n_init=10, max_iter=300,
                 tol=1e-4, random_state=None, copy_x=True):
        super().__init__(n_clusters=n_clusters, init=init, n_init=n_init, max_iter=max_iter,
                         tol=tol, random_state=random_state, copy_x=copy_x)

    def fit(self, X, y=None, sample_weight=None):
        X_normalized = normalize(X, norm='l2')
        self.distance_metric = 'cosine'
        return super().fit(X_normalized, y=y, sample_weight=sample_weight)

# 예제 데이터 생성
X = np.random.rand(100, 5)

# 코사인 거리를 기반으로 K-Means 클러스터링 수행
kmeans = CosineKMeans(n_clusters=3)
labels = kmeans.fit_predict(X)

print(labels)
print(kmeans.cluster_centers_)
```
